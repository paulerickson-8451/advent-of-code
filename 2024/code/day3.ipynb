```python
# read the entire file into a multi-line string
def read_data(file_path):
    with open(file_path, 'r') as file:
        data = file.read()
    return data

file_path = '../data/day-3.txt'
data = read_data(file_path)
```

```python
import re

def extract_and_multiply(memory):
    # Regular expression to match valid mul(X,Y) instructions
    pattern = r'mul\(\d{1,3},\d{1,3}\)'
    
    # Find all matches in the memory string
    matches = re.findall(pattern, memory)
    
    total_sum = 0
    for match in matches:
        # Extract the numbers from the match
        numbers = re.findall(r'\d{1,3}', match)
        x, y = int(numbers[0]), int(numbers[1])
        total_sum += x * y
    
    return total_sum

# Calculate the sum of all valid mul instructions
result = extract_and_multiply(data)
print("Part 1:", result)  # Output should be 161
```

```python
def extract_and_multiply_with_conditions(data):
    # Regular expression to match valid mul(X,Y) instructions
    pattern = r'mul\(\d{1,3},\d{1,3}\)'
    
    # Find all matches in the memory string
    matches = re.findall(pattern, data)
    
    # Initialize the enabled state to True
    enabled = True
    
    total_sum = 0
    for match in matches:
        # Check for do() and don't() instructions before the current match
        do_index = data.rfind('do()', 0, data.find(match))
        dont_index = data.rfind("don't()", 0, data.find(match))
        
        # Update the enabled state based on the most recent instruction
        if do_index > dont_index:
            print("do() enabled")
            enabled = True
        elif dont_index > do_index:
            print("don't() enabled")
            enabled = False
        
        if enabled:
            # Extract the numbers from the match
            numbers = re.findall(r'\d{1,3}', match)
            x, y = int(numbers[0]), int(numbers[1])
            total_sum += x * y
    
    return total_sum

# Calculate the sum of all valid and enabled mul instructions
result_with_conditions = extract_and_multiply_with_conditions(data)
print("Part 2:", result_with_conditions)
```
